```python
df1 = df[['comment_clean', 'sentiment']]
df1
```

```python
df1 = df1.dropna()
df1 = df1.reset_index(drop=True)
df1
```

```python
train_size = 0.7
val_size = 0.15
test_size = 0.15

train_df, remaining_df = train_test_split(df1, test_size=(test_size + val_size), random_state=42)

# Chia tập còn lại thành tập kiểm tra và tập validation
test_df, val_df = train_test_split(remaining_df, test_size=(val_size / (test_size + val_size)), random_state=42)

print("Tập huấn luyện:")
print(train_df)
print("\nTập kiểm tra:")
print(test_df)
print("\nTập validation:")
print(val_df)

```
